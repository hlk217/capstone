### Handle json files to extract the label and count the label frequency.

````python
import json
from pandas.io.json import json_normalize
import numpy as np 
import pandas as pd
import os

data_type = "train"
data = json.load(open('input/' + data_type + '.json'))
labels = json_normalize(data["annotations"])
labels = labels.set_index("imageId")
labelCounts = {}
for i, value in enumerate(labels.values):
    val = value.tolist()[0]
    for j in val:
        if j in labelCounts:
            labelCounts[j] = labelCounts[j] + 1
        else:
            labelCounts[j] = 1   
````
